```python
students = {}

subjects = ["Math", "Physics", "Chemistry", "English"]

pass_marks = {
    "Math": 40,
    "Physics": 35,
    "Chemistry": 35,
    "English": 30
}

def add_student():
    roll = input("Enter Roll Number: ")

    if roll in students:
        print("Student already exists!")
        return

    name = input("Enter Student Name: ")

    students[roll] = {
        "name": name,
        "marks": {}
    }

    print("Student added successfully!\n")

def enter_marks():
    roll = input("Enter Roll Number: ")

    if roll not in students:
        print("Student not found!\n")
        return

    for subject in subjects:
        while True:
            try:
                mark = float(input(f"Enter marks in {subject} (0-100): "))

                if 0 <= mark <= 100:
                    students[roll]["marks"][subject] = mark
                    break
                else:
                    print("Marks must be between 0 and 100.")

            except:
                print("Invalid input.")

    print("Marks entered successfully!\n")

def calculate_average(roll):
    marks = students[roll]["marks"]

    if len(marks) == 0:
        return 0

    total = 0

    for m in marks.values():
        total += m

    return total / len(marks)

def get_grade(avg):

    if avg >= 90:
        return "A+"

    elif avg >= 80:
        return "A"

    elif avg >= 70:
        return "B"

    elif avg >= 60:
        return "C"

    elif avg >= 50:
        return "D"

    else:
        return "F"

def pass_fail_status(roll):

    marks = students[roll]["marks"]

    for subject in subjects:

        if subject not in marks:
            return "Marks Not Entered"

        if marks[subject] < pass_marks[subject]:
            return "FAIL"

    return "PASS"

def student_report():

    roll = input("Enter Roll Number: ")

    if roll not in students:
        print("Student not found!\n")
        return

    print("\n===== STUDENT REPORT =====")

    print("Roll No :", roll)
    print("Name    :", students[roll]["name"])

    for subject, mark in students[roll]["marks"].items():
        print(subject, ":", mark)

    avg = calculate_average(roll)

    print("Average :", round(avg, 2))
    print("Grade   :", get_grade(avg))
    print("Status  :", pass_fail_status(roll))
    print()

def subject_toppers():

    print("\n===== SUBJECT TOPPERS =====")

    for subject in subjects:

        topper_name = ""
        highest = -1

        for roll in students:

            marks = students[roll]["marks"]

            if subject in marks:

                if marks[subject] > highest:
                    highest = marks[subject]
                    topper_name = students[roll]["name"]

        if highest != -1:
            print(f"{subject}: {topper_name} ({highest})")

    print()

def overall_ranking():

    ranking = []

    for roll in students:
        avg = calculate_average(roll)

        ranking.append([
            roll,
            students[roll]["name"],
            avg
        ])

    n = len(ranking)

    for i in range(n):
        for j in range(0, n-i-1):

            if ranking[j][2] < ranking[j+1][2]:

                temp = ranking[j]
                ranking[j] = ranking[j+1]
                ranking[j+1] = temp

    print("\n===== OVERALL RANKING =====")

    rank = 1

    for student in ranking:
        print(
            rank,
            ".",
            student[1],
            "- Avg:",
            round(student[2], 2)
        )

        rank += 1

    print()

def summary_report():

    total_students = len(students)

    passed = 0
    failed = 0

    for roll in students:

        status = pass_fail_status(roll)

        if status == "PASS":
            passed += 1

        elif status == "FAIL":
            failed += 1

    print("\n===== SUMMARY REPORT =====")
    print("Total Students :", total_students)
    print("Passed         :", passed)
    print("Failed         :", failed)

    if total_students > 0:
        print(
            "Pass Percentage:",
            round((passed / total_students) * 100, 2),
            "%"
        )

    print()

def menu():

    while True:

        print("========== STUDENT PERFORMANCE ANALYZER ==========")
        print("1. Add Student")
        print("2. Enter Marks")
        print("3. Student Report")
        print("4. Subject Wise Toppers")
        print("5. Overall Ranking")
        print("6. Summary Report")
        print("7. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            add_student()

        elif choice == "2":
            enter_marks()

        elif choice == "3":
            student_report()

        elif choice == "4":
            subject_toppers()

        elif choice == "5":
            overall_ranking()

        elif choice == "6":
            summary_report()

        elif choice == "7":
            print("Thank You!")
            break

        else:
            print("Invalid Choice!\n")

menu()
```
