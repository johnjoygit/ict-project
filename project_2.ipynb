```python
def play_game():

    score = 0
    decisions = []

    print("\nWELCOME TO THE LOST ISLAND ADVENTURE")
    print("You wake up on a mysterious island after a shipwreck.")

    choice1 = input("\nDo you go into the FOREST or towards the BEACH? ").lower()

    decisions.append(choice1)

    if choice1 == "forest":

        score += 10

        print("\nYou enter the forest and find a strange cave.")

        choice2 = input("Do you ENTER the cave or WALK away? ").lower()

        decisions.append(choice2)

        if choice2 == "enter":

            score += 20

            print("\nInside the cave you find hidden treasure.")

            choice3 = input("Do you TAKE the treasure or LEAVE it? ").lower()

            decisions.append(choice3)

            if choice3 == "take":

                score += 30

                print("\nYou become rich and escape the island.")
                ending = "Treasure Ending"

            else:

                score += 10

                print("\nYou leave the treasure and gain wisdom.")
                ending = "Wise Explorer Ending"

        else:

            score -= 5

            print("\nYou get lost in the forest.")

            ending = "Lost Ending"

    elif choice1 == "beach":

        score += 5

        print("\nYou walk along the beach and find a boat.")

        choice2 = input("Do you REPAIR the boat or SEARCH the island? ").lower()

        decisions.append(choice2)

        if choice2 == "repair":

            score += 20

            print("\nYou successfully repair the boat.")

            choice3 = input("Do you SAIL immediately or REST first? ").lower()

            decisions.append(choice3)

            if choice3 == "sail":

                score += 25

                print("\nYou safely return home.")
                ending = "Hero Ending"

            else:

                score -= 10

                print("\nA storm destroys the boat while you rest.")
                ending = "Storm Ending"

        else:

            score += 10

            print("\nYou discover ancient ruins.")

            choice3 = input("Do you EXPLORE them or IGNORE them? ").lower()

            decisions.append(choice3)

            if choice3 == "explore":

                score += 20

                print("\nYou discover valuable knowledge.")
                ending = "Explorer Ending"

            else:

                score -= 5

                print("\nYou miss an important opportunity.")
                ending = "Missed Chance Ending"

    else:

        print("\nInvalid choice.")
        ending = "Game Over"

    print("\n========== GAME SUMMARY ==========")
    print("Decisions Made:")

    for i in range(len(decisions)):
        print(f"{i+1}. {decisions[i]}")

    print("Final Score:", score)
    print("Ending:", ending)
    print("==================================\n")


def menu():

    while True:

        print("===== TEXT BASED DECISION GAME =====")
        print("1. Play Game")
        print("2. Exit")

        choice = input("Enter Choice: ")

        if choice == "1":
            play_game()

        elif choice == "2":
            print("Thank you for playing!")
            break

        else:
            print("Invalid Choice\n")


menu()
```
